##**Indexing numeric arrays in MATLAB**

- **Indexing** is a way to access and manipulate a subset of elements in an array
  - simular to pulling an index in Python
  - allows you to use for another task or rewrite/update elements
- to index a vector, you use **(position) **
  - vectors in MATLAB do not start at position 0 like Python does- rather it starts counting at 1

```
vec = [ 1, 5, 8, 2, 0, 10]

out = vec(4)

```
  - typing *out* would print the fourth position of the vector, which in this case is 2
- to get multiple values, you can use **(start position: end position)**

```
out = vec(3:5)

```
  - typing *out* in this case would print [8, 2, 0]
- to get non-consecutive values, you can use **([position 1, position 2, etc.])**

```
out = vec([1,3,6])

#using both tactics...

out = vec([2:3, 6:8])
```

- **end** is similar to the (-1) index in Python; *vec(end)* or *vec(3:end)*

- to **replace** elements in a vector, you use the same technique:

```
vec(4:6) = [12, 6, 1]
```

- you can also use these techniques to carry information from one vector to another and column vectors

##**Conditional Statements in MATLAB**

- **Conditional statements** let programs make decisions based on whether a condition is *true*

####*Syntax*
```
if the condition is true
  run code in this section;
else
  run this section instead;
end

#OR

if condition is true
  run code in this section
elseif this other condition is true
  run this other section instead
.
.
.
else
  run this section instead;
end
```
We can have the following example:

```
corr_pass = 2345
entered_pass = 2345

if entered_pass == corr_pass
  disp('You entered the correct passcode!')
else
  disp('You entered the wrong passcode!')
end
```

- like in python, '==' tests if the two variables are equal
  - you can use other operational conditions like the following:
    - ~= : is *not* equal to (the same as != in Python)
    - > : greater than
    - >= : greater than or equal to
    - < : less than
    - <= : less than or equal to
    - strcmp(a,b) : string *a* is equal to string *b*
  

####**MATLAB *for* and *while* loop syntax**

- Loops allow you to repeat sections of code automatically
  - *for loops* repeat a specific number of times
  - *while Loops* repeat until a specific condition is met
```
for counter = low:increment:high
  %do some work
end
#for example:
sum = 0;
for i=5:1:30
  sum = sum+i;
end
######
while condition
  %do some work; condition must change inside loop
end
#for example:
n = 997992+2;
while ~isprime(n)
  n=n+1;
end
  disp(['result= ',num2str(n)]);
```
- if we need the loop to stop at a certain condition, we can use **break;** to force MATLAB to exit the loop
- to terminate a never-ending loop, you can use *ctrl + c*
- you can also loop through vectors + do nested loops:
```
#vector loop example
v = [1,4,2,5,6,7];
for i = 1:1:length(v)
  v(i)
end
#nested for loop example
v = [1,4,2;5,6,7;1,3,8];
[r,c] = size(v);
for j = 1:1:r
  for i=1:1:c
    v(j,i)
  end
end
```

- to test conditions *simultaneously*, use **'&&'** (and), or you use **'||'** (or)
